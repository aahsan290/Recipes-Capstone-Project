```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
recipes = pd.read_csv('C:\\Users\\aahsa\\Downloads\\recipes.csv')
```

```python
recipes.info()
```

```python
recipes.head()
```

```python
missing_values = recipes.isnull()
print(missing_values)
```

```python
import numpy as np
recipes.replace(False, np.nan, inplace=True)
```

```python
# Check for missing values
missing_summary = recipes.isnull().sum()
print(missing_summary)
```

```python
recipes.drop(['DatePublished'], axis=1, inplace=True)
```

```python
recipes.head()
```

```python
# Remove duplicate rows
recipes.drop_duplicates(inplace=True)
```

```python
# Replace 'PT' with an empty string, 'H' with 'hour', and 'M' with 'mins'
recipes['TotalTime'] = recipes['TotalTime'].str.replace('PT', '').str.replace('H', ' hour ').str.replace('M', ' mins')
recipes['PrepTime'] = recipes['PrepTime'].str.replace('PT', '').str.replace('H', ' hour ').str.replace('M', ' mins')
recipes.head()
```

```python
columns = list(recipes.columns)
columns.insert(columns.index('RecipeIngredientParts') + 1, columns.pop(columns.index('RecipeInstructions')))
recipes= recipes[columns]
```

```python
recipes.drop('AuthorId',axis=1, inplace=True)
recipes.drop('RecipeYield',axis=1, inplace=True)
recipes.drop('CookTime',axis=1, inplace=True) 
recipes.drop('RecipeId', axis=1, inplace=True)
```

```python
recipes.head()
```

```python
recipes.head()
```

```python
# Combine columns into one "nutrition" column with labels
recipes['nutrition'] = recipes.apply(lambda row: f"Saturated_Fat: {row['SaturatedFatContent']}, Cholestrol: {row['CholesterolContent']}, Sodium: {row['SodiumContent']}, Carbs: {row['CarbohydrateContent']}, Sodium: {row['SodiumContent']}, Fiber: {row['FiberContent']}, Sugar: {row['SugarContent']}, Protein: {row['ProteinContent']}, Fat: {['FatContent']} ", axis=1)
```

```python
recipes.drop(['SaturatedFatContent',	'CholesterolContent', 'SodiumContent', 'CarbohydrateContent', 'FiberContent', 'SugarContent', 'ProteinContent','FatContent'], axis=1, inplace=True)
```

```python
recipes.head()
```

```python
recipes['CombinedIngredients'] = recipes['RecipeIngredientQuantities'] + ' ' + recipes['RecipeIngredientParts']
```

```python
recipes.drop(['RecipeIngredientQuantities', 'RecipeIngredientParts'], axis=1, inplace=True)
```

```python
recipes.head()
```

```python
# Move the "AuthorName" column to the end 
columns = list(recipes.columns) 
columns.append(columns.pop(columns.index('AuthorName')))
recipes = recipes[columns] 
```

```python
recipes.fillna(1.0, inplace=True)
```

```python
# Check for missing values
missing_summary = recipes.isnull().sum()
print(missing_summary)
```
